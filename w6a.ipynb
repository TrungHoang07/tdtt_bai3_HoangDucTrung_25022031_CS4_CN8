```python
#bài 4
s = input().split()
d = {}
for i in s:
    key,value = i.split(":")
    if key not in d:
        d[key] = [value]
    else:
        d[key].append(value)
print(d)
```

```python
#bài 5
l = list(map(int, input().split()))
d = {"positives": 0, "negatives":0, "zeros":0}
for i in l:
    if i > 0:
        d["positives"] += 1
    elif i < 0:
        d["negatives"] += 1
    elif i == 0:
        d["zeros"] += 1
print(d)
```

```python
#bài 9
a = eval(input())
b = eval(input())
c = {}
for i in a:
    if i in b:
        c[i] = a[i] + b[i]
    else:
        c[i] = a[i]
for i in b:
    if i not in c:
        c[i] = b[i]
print(c)
```

```python
#bài 10
l = list(map(int, input().split()))
k = int(input())
a = set()
for i in range(len(l)):
    for j in range(i + 1, len(l)):
        if l[i] + l[j] == k:
            a.add((l[i], l[j]))
result = sorted(list(a))
print(result)
```

```python
#bài 17
row = int(input())
col = row
n = []
s = 0
for i in range(row):
    n.append([])
    for j in range(1,col+1):
        s += 1
        n[i].append(s)
for i in range(len(n)):
    print(n[i][i], end = " ")
print()
k = len(n)
for i in range(len(n)):
    k -= 1
    print(n[i][k], end = " ")
```

```python
#bài 20
def find(l,k):
    if k in l:
        return l.index(k) +1
    return -1
l = list(map(int, input().split()))
k = int(input())
print(find(l,k))
```

```python
#bài 21
s = input().split()
d = {}
for i,v in enumerate(s):
    if v not in d:
        d[v] = [s.index(v)]
    else:
        d[v].append(i)
for k,v in d.items():
    d[k] = tuple(v)
print(d)
```
