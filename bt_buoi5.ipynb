```python
def find(l,k):
    s = 0
    for i in range(len(l)):
        if l[i] == k:
            s+=1
            return i+1
            break
    if s == 0:
        return -1
lst = list(map(int, input().split()))
n = int(input())
print(find(lst,n))

```

```python
def cal(a,b,s):
    if "+" == s:
        return a+b
    if "-" == s:
        return a-b
    if "*" == s:
        return a*b
    if "/" == s:
        return a/b
a_,op,b_ = input().split()
a_ = float(a_)
b_ = float(b_)
print(f"{(cal(a_,b_,op)):.2f}")

```

```python
def hamming(x, y):
    xor = x ^ y
    distance = bin(xor).count('1')
    return distance
a,b = map(int,(input()).split())
print(hamming(a,b))

```

```python
def check(s, t):
    map_s_t = {}
    map_t_s = {}
    for i in range(len(s)):
        char_s = s[i]
        char_t = t[i]
        if char_s in map_s_t:
            if map_s_t[char_s] != char_t:
                return False
        else:
            if char_t in map_t_s:
                return False
            map_s_t[char_s] = char_t
            map_t_s[char_t] = char_s
    return True
m,n = input().split()
print(check(m,n))
```
