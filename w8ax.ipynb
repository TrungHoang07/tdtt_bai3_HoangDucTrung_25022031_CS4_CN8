```python
#B1
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        self.pi = 3.14

    def area(self):
        return 2 * self.pi * (self.radius ** 2) + 2 * self.pi * self.radius * self.height

    def volume(self):
        return self.pi * (self.radius ** 2) * self.height

r, h = map(float, input().split())

hinh_tru = Cylinder(r, h)

dt = hinh_tru.area()
v = hinh_tru.volume()

print(f"{dt:.2f} {v:.2f}")
```

```python
#B2
class Date():
    def __init__(self,d,m,y):
        self.day = d
        self.month = m
        self.year = y
    def check_val(self):
        days_31 = [1,3,5,7,8,10,12]
        days_30 = [4,6,9,11]
        if self.month == 2:
            if self.year % 4 == 0 and self.year % 100 != 0:
                if  1 <= self.day <=29:
                    return "VALID"
                else: return "INVALID"
            else:
                if  1 <= self.day <=28:
                    return "VALID"
                else: return "INVALID"
        else:
            if 1 <= self.month <= 12:
                if self.month in days_31:
                    if 1<= self.day <= 31:
                        return "VALID"
                    else: return "INVALID"
                elif self.month in days_30:
                    if 1<= self.day <= 30:
                        return "VALID"
                    else: return "INVALID"
            else:
                return "INVALID"
d,m,y = map(float, input().split())
dinh_dang = Date(d,m,y)
print(dinh_dang.check_val())
```

```python
#B3
class Point():
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def value(self):
        dis = (self.x*self.x + self.y*self.y)
        return dis**0.5
    def check(self):
        if self.x == 0 and self.y == 0 :
            return "M thuoc goc toa do"
        elif self.x != 0 and self.y == 0:
            return "M nam tren truc hoanh"
        elif self.x == 0 and self.y != 0:
            return "M nam tren truc tung"
x,y = map(float, input().split())
m = Point(x,y)
print(m.check())
print(f"{m.value():.2f}")
```

```python
#B4
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def set_numbers(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        return self.num1 + self.num2

    def subtract(self):
        return self.num1 - self.num2

    def product(self):
        return self.num1 * self.num2

    def divide(self):
        return self.num1 / self.num2 if self.num2 != 0 else "Loi chia cho 0"

    def power(self):
        return self.num1 ** self.num2

    def mod(self):
        return self.num1 % self.num2 if self.num2 != 0 else "Loi chia cho 0"

a = float(input("Nhap so thu nhat: "))
b = float(input("Nhap so thu hai: "))
calc = Calculator(a, b)

while True:
    print("""\n--- MENU PHEP TINH ---
         1. Cong (add)      2. Tru (subtract)
         3. Nhan (product)   4. Chia (divide)
         5. Luy thua (power) 6. Chia du (mod)
         7. Cap nhat 2 so moi (set_numbers)""")

    chon = input("Chon phep tinh (1-7): ")

    if chon == '1': print(f"Ket qua: {calc.add()}")
    elif chon == '2': print(f"Ket qua: {calc.subtract()}")
    elif chon == '3': print(f"Ket qua: {calc.product()}")
    elif chon == '4': print(f"Ket qua: {calc.divide()}")
    elif chon == '5': print(f"Ket qua: {calc.power()}")
    elif chon == '6': print(f"Ket qua: {calc.mod()}")
    elif chon == '7':
        mới_a = float(input("Nhap so thu nhat moi: "))
        mới_b = float(input("Nhap so thu hai moi: "))
        calc.set_numbers(mới_a, mới_b)

    thoat = input("Ban co muon thoat? (yes/YES/Yes de thoat): ")
    if thoat.lower() == "yes":
        print("Tam biet!")
        break
```

```python
#B5
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, name, price):
        item = {"name": name, "price": price}
        self.items.append(item)
        print(f"Da them '{name}' vao gio hang.")

    def remove_item(self, name):
        found = False
        for item in self.items:
            if item["name"] == name:
                self.items.remove(item)
                print(f"Da xoa '{name}' khoi gio hang.")
                found = True
                break
        if not found:
            print(f"Khong tim thay san pham '{name}' de xoa.")

    def is_empty(self):
        return len(self.items) == 0

    def calculate_total(self):
        total = sum(item["price"] for item in self.items)
        return total

    def display_cart(self):
        if self.is_empty():
            print("Gio hang dang trong.")
        else:
            print("--- Danh sach san pham ---")
            for i, item in enumerate(self.items, 1):
                print(f"{i}. {item['name']} - Gia: {item['price']}")

    def clear_cart(self):
        self.items.clear()
        print("Da xoa toan bo gio hang.")

cart = ShoppingCart()

while True:
    print("\n========= MENU GIO HANG =========")
    print("2. Them san pham")
    print("3. Xoa san pham theo ten")
    print("4. Kiem tra gio hang rong")
    print("5. Tinh tong tien")
    print("6. Hien thi tat ca san pham")
    print("7. Xoa sach gio hang")

    choice = input("Nhap lua chon cua ban (2-7): ")

    if choice == '2':
        name = input("Nhap ten san pham: ")
        price = float(input("Nhap gia san pham: "))
        cart.add_item(name, price)
    elif choice == '3':
        name = input("Nhap ten san pham can xoa: ")
        cart.remove_item(name)
    elif choice == '4':
        if cart.is_empty():
            print("Gio hang dang rong.")
        else:
            print("Gio hang hien dang co san pham.")
    elif choice == '5':
        print(f"Tong tien trong gio hang: {cart.calculate_total()}")
    elif choice == '6':
        cart.display_cart()
    elif choice == '7':
        cart.clear_cart()
    else:
        print("Lua chon khong hop le.")

    exit_prompt = input("Co muon thoat khoi chuong trinh khong? (Nhap 'yes' de thoat): ")
    if exit_prompt.lower() == "yes":
        print("Cam on ban da su dung dich vu!")
        break
```

```python
#B6
class Rectangle():
    def __init__(self,w,h,k):
        self.w = w
        self.h = h
        self.k = k
    def area(self):
        return ( self.w * self.h ) * (self.k**2)
    def peri(self):
        return ( self.w + self.h ) * 2 * self.k

w,h,k = map(float, input().split())
rec = Rectangle(w,h,k)
print(f"{rec.area():.2f} {rec.peri():.2f}")
```

```python
#B7
class Fraction():
    def __init__(self,a,b,c,d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    def add(self):
        if b == d:
            return f"{a+c}/{b}"
        else:
            return f"{a*c + b*d}/{b*d}"
    def sub(self):
        if b == d:
            return f"{a-c}/{b}"
        else:
            return f"{a*c - b*d}/{b*d}"
    def mul(self):
        return f"{a*c}/{b*d}"
    def div(self):
        return f"{a*d}/{b*c}"

a,b,op,c,d = map(str, input().split())
a = int(a)
b = int(b)
c = int(c)
d = int(d)
frac = Fraction(a,b,c,d)
def uscln(s1,s2):
        if (s2 == 0):
            return s1
        return uscln(s2, s1 % s2)
if op == "+":
    tong = str(frac.add())
    o = tong.index("/")
    s1 = int(tong[0:o])
    s2 = int(tong[(o+1)::])
    print(f"{int(s1/uscln(s1,s2))}/{int(s2/uscln(s1,s2))}")
elif op == "-":
    hieu = str(frac.sub())
    o = hieu.index("/")
    s1 = int(hieu[0:o])
    s2 = int(hieu[(o+1)::])
    print(f"{int(s1/uscln(s1,s2))}/{int(s2/uscln(s1,s2))}")
elif op == "*":
    tich = str(frac.mul())
    o = tich.index("/")
    s1 = int(tich[0:o])
    s2 = int(tich[(o+1)::])
    print(f"{int(s1/uscln(s1,s2))}/{int(s2/uscln(s1,s2))}")
elif op == "/":
    thuong = str(frac.div())
    o = thuong.index("/")
    s1 = int(thuong[0:o])
    s2 = int(thuong[(o+1)::])
    print(f"{int(s1/uscln(s1,s2))}/{int(s2/uscln(s1,s2))}")

```

```python
#B8
class BankAccount:
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance
    def deposit(self,x):
        self.balance += x
    def withdraw(self,x):
        if self.balance < x:
            self.balance = self.balance
        else :
            self.balance -= x
if __name__ == '__main__':
    a = input().split()
    bank = BankAccount(a[0],float(a[1]))
    n = int(input())
    while n > 0:
        giao_dich = input().split()
        if giao_dich[0] == 'DEPOSIT':
            bank.deposit(float(giao_dich[1]))
        elif giao_dich[0] == 'WITHDRAW':
            bank.withdraw(float(giao_dich[1]))
        n -= 1
    print(f'{bank.balance:.2f}')
```

```python
#B9
class Student:
    def __init__(self,name):
        self.name = name
        self.sum = 0
        self.cnt = 0
    def add(self,subject,score):
        self.score = score
    def tong(self):
        self.sum += self.score
        self.cnt += 1
    def avg(self):
        result = self.sum / self.cnt
        return result
    def rank(self):
        if self.avg() >= 8:
            return 'Excellent'
        elif self.avg() >= 6.5:
            return 'Good'
        elif self.avg() >= 5:
            return 'Average'
        else:
            return 'Poor'
if __name__ == '__main__':
    name = input()
    n = int(input())
    hs = Student(name)
    while n > 0:
        subject_score = input().split()
        hs.add(subject_score[0],float(subject_score[1]))
        hs.tong()
        n -= 1
    print(name,f'{hs.avg():.2f}',hs.rank())

```

```python
#B10
class Book:
    def __init__(self,title,author,year):
        self.title = title
        self.author = author
        self.year = year
class Library:
    def __init__(self):
        self.book = []
    def add(self,book):
        self.book.append(book)
    def count_by_author(self,name):
        cnt = 0
        for book in self.book:
            if book.author == name:
                cnt += 1
        return cnt
    def find_by_year(self,year):
        cnt = 0
        for book in self.book:
            if book.year == year:
                cnt += 1
        return cnt
if __name__ == '__main__':
    n = int(input())
    lib = Library()
    while n > 0:
        thao_tac = input().split()
        if thao_tac[0] == 'ADD':
            b = thao_tac[1].split(';')
            book = Book(b[0],b[1],b[2])
            lib.add(book)
        elif thao_tac[0] == 'COUNT':
            print(lib.count_by_author(thao_tac[1]))
        elif thao_tac[0] == 'COUNTYEAR':
            print(lib.find_by_year(thao_tac[1]))
        n -= 1
```

```python
#B11
class Angle :
    def __init__(self,a,b,c,d):
        self.ABx = b[0] - a[0]
        self.ABy = b[1] - a[1]
        self.ABz = b[2] - a[2]
        self.ACx = c[0] - a[0]
        self.ACy = c[1] - a[1]
        self.ACz = c[2] - a[2]
        self.BCx = c[0] - b[0]
        self.BCy = c[1] - b[1]
        self.BCz = c[2] - b[2]
        self.BDx = d[0] - b[0]
        self.BDy = d[1] - b[1]
        self.BDz = d[2] - b[2]
    def vector(self):
        self.ABC = [self.ABy * self.ACz - self.ABz * self.ACy, self.ABz * self.ACx - self.ABx * self.ACz, self.ABx * self.ACy - self.ABy * self.ACx]
        self.BCD = [self.BCy * self.BDz - self.BCz * self.BDy, self.BCz * self.BDx - self.BCx * self.BDz, self.BCx * self.BDy - self.BCy * self.BDx]
    def cos(self):
        tich_vo_huong = self.ABC[0] * self.BCD[0] + self.ABC[1] * self.BCD[1] + self.ABC[2] * self.BCD[2]
        tich_do_dai = ((self.ABC[0] ** 2 + self.ABC[1] ** 2 + self.ABC[2] ** 2) ** 0.5) *((self.BCD[0] ** 2 + self.BCD[1] ** 2 + self.BCD[2] ** 2) ** 0.5)
        cos = abs(tich_vo_huong/tich_do_dai)
        return cos
if __name__ == '__main__' :
    A = list(map(float,input().split()))
    B = list(map(float,input().split()))
    C = list(map(float,input().split()))
    D = list(map(float,input().split()))
    my_angle = Angle(A,B,C,D)
    my_angle.vector()
    my_angle = acos(my_angle.cos())
    my_angle = degrees(my_angle)
    print(f'{my_angle:.2f}')
```

```python
#B12
class Complex:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def pro(self,other):
        self.thuc = self.a * other.a - self.b * other.b
        self.ao = self.a * other.b + self.b * other.a
if __name__ == '__main__':
    a = list(map(int,input().split()))
    b = list(map(int,input().split()))
    my_complex = Complex(a[0],a[1])
    other_complex = Complex(b[0],b[1])
    my_complex.pro(other_complex)
    print(my_complex.thuc,my_complex.ao)
```

```python
#B13
class Triangle:
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
    def getArea(self):
        if self.a + self.b <= self.c or self.b + self.c <= self.a or self.a + self.c <= self.b:
            return -1
        self.p = (self.a + self.b + self.c)/2
        return (self.p * (self.p -self.a) * (self.p - self.b) * (self.p -self.c)) ** 0.5
if __name__ == '__main__':
    my_tri = Triangle(*map(int,input().split()))
    if my_tri.getArea() == -1 :
        print('invalid')
    else:
        print(f'{my_tri.getArea():.2f}')
```
